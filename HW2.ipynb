```python
import numpy as np # linear algebra
import scipy as sp 
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import seaborn as sns

# Visualization
import matplotlib.pyplot as plt
```

```python
x = np.arange(10)
predict_y = x
actual_y = x-(np.random.rand(10)*2-1)

plt.scatter(x,actual_y,color ='red',label='actual')
plt.plot(x,predict_y,label='predict')

plt.xlabel('x value')
plt.ylabel('y value')

plt.legend()
plt.title('Data analysis')
plt.show()
```
